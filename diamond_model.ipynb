```python
import torch
import pandas as pd
import matplotlib
import numpy as np
```

```python
import os
```

```python
from PIL import Image
```

```python
blank_image = open('diamond_dataset/blank.png', 'rb').read()
```

```python
diamonds_str = 'diamond_dataset/Diamonds/images/'
for type_name in os.listdir(diamonds_str):
    print(type_name)
    for image in os.listdir(str(diamonds_str + type_name)):
        if str(diamonds_str + type_name + '/' + image) != 'diamond_dataset/Diamonds/images/round/.ipynb_checkpoints':
            current_image = open(str(diamonds_str + type_name + '/' + image), 'rb').read()
            if current_image == blank_image:
                os.remove(str(diamonds_str + type_name + '/' + image))
```

```python
diamonds2_str = 'diamond_dataset/Diamonds2/images/'
for type_name in os.listdir(diamonds2_str):
    print(type_name)
    for image in os.listdir(str(diamonds2_str + type_name)):
        if str(diamonds2_str + type_name + '/' + image) != 'diamond_dataset/Diamonds2/images/round/.ipynb_checkpoints':
            current_image = open(str(diamonds2_str + type_name + '/' + image), 'rb').read()
            if current_image == blank_image:
                os.remove(str(diamonds2_str + type_name + '/' + image))
```

```python
bad_images = []
diamonds_str = 'diamond_dataset/Diamonds/images/'
for type_name in os.listdir(diamonds_str):
    print(type_name)
    for image in os.listdir(str(diamonds_str + type_name)):
        if str(diamonds_str + type_name + '/' + image) != 'diamond_dataset/Diamonds/images/round/.ipynb_checkpoints':
            current_image = Image.open(str(diamonds_str + type_name + '/' + image))
            if current_image.width != 300 or current_image.height != 300:
                print("Images inconsistent")
                bad_images.append(str(diamonds_str + type_name + '/' + image))
```
